<h4>Importing the Libraries</h4>

```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

<h4>Seaborn's built-in datasets</h4>

```python
sns.get_dataset_names()
```

```python
# Total bill vs Tip dataset
tips = sns.load_dataset('tips')
tips.head()
```

```python
# Apply the default theme
sns.set_theme()
```

```python
# Visualize the tips dataset
sns.relplot(
    data=tips,
    x="total_bill", y="tip", col="time",
    hue="smoker", style="smoker", size="size",
)
```

```python
# Load iris dataset
iris = sns.load_dataset('iris')
iris.head()
```

<h4>Scatter plot of iris dataset</h4>

```python
sns.scatterplot(x='sepal_length', y='petal_length', hue='species', data=iris)
```

```python
# Load titanic dataset
titanic = sns.load_dataset('titanic')
titanic.head()
```

<h4>Count Plot of titanic dataset</h4>

```python
sns.countplot(x='class', data=titanic)
```

```python
sns.countplot(x='survived', data=titanic)
```

<h4>Bar chart of titanic dataset</h4>

```python
sns.barplot(x='sex', y='survived', hue='class', data=titanic)
```

```python
# house price dataset
from sklearn.datasets import load_diabetes
diabetes = load_diabetes()
diabetes_data = pd.DataFrame(diabetes.data, columns=diabetes.feature_names)
diabetes_data['SUGAR_LEVEL'] = diabetes.target
```

```python
diabetes_data.head()
```

<h4>Distribution Plot of diabetes dataset</h4>

```python
sns.distplot(diabetes_data['SUGAR_LEVEL'])
```

<h4>Heatmap of diabetes dataset</h4>

```python
correlation = diabetes_data.corr()
```

```python
plt.figure(figsize=(10, 10))
sns.heatmap(correlation, cbar=True, square=True, fmt='.1f', annot=True, annot_kws={'size': 8}, cmap='Blues')
```
