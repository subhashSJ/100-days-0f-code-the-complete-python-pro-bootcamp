<h4>Import Dependencies</h4>

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestRegressor
from sklearn import metrics
```

<h4>Data Collection and Processing</h4>

```python
# Loading csv data into pandas data frame
df = pd.read_csv('gold_price_data.csv')
df.head()
```

```python
df.shape
```

```python
df.info()
```

```python
# Statastical measures of the data
df.describe()
```

```python
# Check number of missing values
df.isnull().sum()
```

Correlation

```python
correlation = df.drop(columns='Date', axis=1).corr()
```

```python
# Constructing heatmap to understand correlation
plt.figure(figsize=(8,8))
sns.heatmap(correlation, cbar=True, square=True, fmt='.1f', annot=True, annot_kws={'size':8}, cmap='Blues')
```

```python
# Correlation values for GLD
correlation['GLD']
```

```python
# Checking distribution of GLd values
sns.distplot(df['GLD'], color='green')
```

<h4>Splitting features and target</h4>

```python
X=df.drop(columns=['Date', 'GLD'], axis=1)
Y=df['GLD']
```

```python
print(X)
```

```python
print(Y)
```

<h4>Splitting training and testing data</h4>

```python
X_train, X_test, Y_train, Y_test = train_test_split(X,Y, test_size=0.2, random_state=2)
```

```python
print(X.shape, X_train.shape, X_test.shape)
```

<h4>Model Training: Random Forest Regressor</h4>

```python
regressor = RandomForestRegressor(n_estimators=100)
```

```python
# Training the model
regressor.fit(X_train, Y_train)
```

</h4>Model Evaluation</h4>

```python
#Prediction on Test data
test_data_prediction = regressor.predict(X_test)
```

```python
# R squared error
error_score = metrics.r2_score(Y_test, test_data_prediction)

print(f'R2 error score: {error_score}')
```

<h4>Compare actual and predicted values</h4>

```python
Y_test = list(Y_test)
```

```python
plt.plot(Y_test, color='blue', label='Actual Value')
plt.plot(test_data_prediction, color='green', label='Predicted Value')
plt.title('Actual Price vs Predicted Price')
plt.xlabel('Number of values')
plt.ylabel('GLD Price')
plt.legend()
plt.show()
```
